```python
import pandas as pd

# Read the CSV file into a DataFrame
data = pd.read_csv("../Alzheimer_s_Disease_and_Healthy_Aging_Data.csv")

# Display the DataFrame
data.head(20)
```

```python
data.columns
```

```python
columns_to_drop = ['StratificationCategoryID1', 'YearEnd', 'LocationAbbr', 'Topic', 'Datasource', 'DataValueTypeID', 'Data_Value_Footnote_Symbol', 'StratificationCategory1', 'ClassID', 'TopicID', 'QuestionID', 'LocationID']
data = data.drop(columns=columns_to_drop)
```

```python
data.columns
```

```python
data.head()
```

```python
columns_to_drop2 = ['RowId']

data = data.drop(columns=columns_to_drop2)

data.head()
```

```python
data.rename(columns={'Stratification1': 'Age Group'}, inplace=True)
data.head()
```

```python
data.drop(columns=['StratificationID1'], inplace=True)
data.head()
```

```python
data.drop(columns=['StratificationID2'], inplace=True)
data.head()
```

```python
data.drop(columns=['Data_Value_Unit'], inplace=True)
data.head()
```

```python
data.drop(columns=['Data_Value_Type'], inplace=True)
data.rename(columns={'YearStart': 'Year'}, inplace=True)
data.head()
```

```python
data.drop(columns=['Latitude', 'Longitude'], inplace=True)
data.head()
```

```python
data.to_csv("../Alzheimer_s_Disease_and_Healthy_Aging_Data_1st_Update.csv", index=False)
```

```python
data.drop(columns=['Data_Value_Alt'], inplace=True)
data.head()
```

```python
data['Gender'] = data['Stratification2'].apply(lambda x: x if x in ['Male', 'Female'] else None)

data.head()
```

```python
data['Race/Ethnicity'] = data['Stratification2'].apply(lambda x: x if x in ['Asian/Pacific Islander', 'White, non-Hispanic', 'Black, non-Hispanic', 'Native Am/Alaskan Native', 'Hispanic'] else None)

data.head()
```

```python
data.drop(columns=['StratificationCategory2', 'StratificationCategoryID2'], inplace=True)
data.head()
```

```python
data['Gender'] = data['Gender'].str.replace('None', 'N/A')
data.head()
```

```python
data.to_csv("../Alzheimers_Data_2nd_Update.csv", index=False)
```

```python
data.rename(columns={'Stratification2': 'Gender or Race/Ethnicity'}, inplace=True)
data.head()
```

```python
data.to_csv("../Alzheimers_Data_3rd_Update.csv", index=False)
```

```python
#Calculate the halfway point
half = len(data) // 2

# Split the data
data1 = data.iloc[:half]
data2 = data.iloc[half:]

# Save the new DataFrames to csv
data1.to_csv('1st_half_data.csv', index=False)
data2.to_csv('2nd_half_data.csv', index=False)
```
